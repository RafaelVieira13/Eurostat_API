```python
# Data Cleaning
data = eurostat.get_data_df('MET_PJANAGGR3')

# Drop unecessarry columns
data = data.drop(columns={'freq'})

# Let's Create a New Column Called 'Time' and another one called 'Value'
data = pd.melt(data, id_vars=['sex','age','metroreg\TIME_PERIOD'], var_name='Time', value_name='Value')

# Let's Drop Missing Values
data = data.dropna()

# Sort The data by 'Time' and 'age'
data.sort_values(by=['Time','age'], ascending=[False,True], inplace=True)

data
```

## Function to Save All The Data And Schedule To Automatically Run The Code 

```python
import eurostat
import time
import schedule
import pandas
import os

def get_and_save_data():
    try:
        print("Getting data...")
        # Let's use the code to get the 'Population by Geo and Time'
        data_geo_time = eurostat.get_data_df('tps00001')

        # Let's store the data in a folder
        folder_path = 'C:/Users/vieirra/Desktop/EUROSTAT_API'

        # Check if the folder exists, if not, create it
        os.makedirs(folder_path, exist_ok=True)

        # Save the data to a CSV file in the folder
        file_path = os.path.join(folder_path, 'population.csv')
        data_geo_time.to_csv(file_path, index=False)

        print(f"Data fetched and saved to {file_path}")
    except Exception as e:
        print(f"Error: {e}")

# Schedule the data download to run every day at 2:00 PM
schedule.every().day.at("09:44").do(get_and_save_data)

# Run the scheduled job continuously
while True:
    schedule.run_pending()
    time.sleep(1)
```
